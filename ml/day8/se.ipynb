```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = {'Name': ['ANSH', 'SAHIL', 'JAYAN', 'ANURAG'], 'Age': [21, 23, 20, 24]}
df = pd.DataFrame(data)

sns.lineplot(x=df.index, y='Age', data=df)
plt.xlabel('Index')
plt.ylabel('Age')
plt.title('Age Line Plot')
plt.show()
```
