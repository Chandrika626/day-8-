```python
import matplotlib.pyplot as plt

x = [10, 20, 30, 40]
y = [20, 25, 35, 55]

plt.plot(x, y)

plt.title("Line Chart")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")

plt.show()
```

```python
import matplotlib.pyplot as plt

days = ['Mon', 'Tue', 'Wed', 'Thu']
sales = [100, 150, 120, 180]

plt.bar(days, sales)

plt.title("Bar Chart")
plt.xlabel("Days")
plt.ylabel("Sales")

plt.show()
```

```python
import matplotlib.pyplot as plt

data = [1,2,2,3,3,3,4,4,5,5,5,6]

plt.hist(data, bins=5)

plt.title("Histogram")
plt.xlabel("Values")
plt.ylabel("Frequency")

plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [2,4,6,8,10]

plt.scatter(x, y)

plt.title("Scatter Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")

plt.show()
```

```python
import matplotlib.pyplot as plt

cars = ['BMW', 'AUDI', 'TESLA', 'FORD']
data = [20, 30, 25, 25]

plt.pie(data, labels=cars, autopct='%1.1f%%')

plt.title("Pie Chart")

plt.show()
```

```python
import matplotlib.pyplot as plt

data = [[10,20,30,40],
        [15,25,35,45],
        [12,22,32,42]]

plt.boxplot(data)

plt.title("Box Plot")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.rand(5,5)

plt.imshow(data, cmap='viridis')

plt.colorbar()

plt.title("Heatmap")

plt.show()
```
